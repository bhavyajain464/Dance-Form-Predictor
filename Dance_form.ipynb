```python
from PIL import Image
import numpy as np
from numpy import asarray
import pandas as pd
import matplotlib.pyplot as plt
import cv2
# matplotlib %inline
c=0
list = []
# dataset = np.array()
import os
for filename in os.listdir(path='/home/bhavya/Downloads/dance_form/train'):
    if filename.endswith(".jpg"):
        path='/home/bhavya/Downloads/dance_form/train/'+filename
        image = Image.open(path)
        list.append(filename)
        data = asarray(image)
        data.shape
        image_resized = cv2.resize(data,(126,139),interpolation=cv2.INTER_NEAREST)
#         print(image_resized.shape)
        image_data = image_resized.reshape(52542)
        if c==0:
            dataset = np.array(image_data)
#             print(dataset)
        else:
            dataset = np.append(dataset,image_data)
#             print(dataset)
        c+=1
        continue
    else:
        continue
```

```python
139*126*3
```

```python
train_images = dataset.reshape(364,52542)
train_images= train_images.astype('float32')
train_images/=255
train_images
```

```python
target = pd.read_csv('train.csv')
for i in range(0,364):
#     print(target['target'][i])
    if(target['target'][i]=='manipuri'):
        target['target'][i]=0
    elif(target['target'][i]=='bharatanatyam'):
        target['target'][i]=1
    elif(target['target'][i]=='odissi'):
        target['target'][i]=2
    elif(target['target'][i]=='kathakali'):
        target['target'][i]=3
    elif(target['target'][i]=='kathak'):
        target['target'][i]=4
    elif(target['target'][i]=='sattriya'):
        target['target'][i]=5
    elif(target['target'][i]=='kuchipudi'):
        target['target'][i]=6
    elif(target['target'][i]=='mohiniyattam'):
        target['target'][i]=7
print(target)
```

```python
target
```

```python
frame = pd.DataFrame(data=list,columns=['Image'])
frame
```

```python
train_label = pd.merge(frame,target,on='Image')['target']
```

```python
from tensorflow import keras
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop
train_label = train_label.to_numpy()
train_labels = keras.utils.to_categorical(train_label,8)
```

```python
model = Sequential()
# model.add(Dense(256,activation='relu',input_shape=(52542,)))
model.add(Dense(54,activation='relu',input_shape=(52542,)))
# model.add(Dense(10,activation='relu',input_shape=(52542,)))
model.add(Dense(8,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer=RMSprop(),metrics=['accuracy'])
```

```python
history = model.fit(train_images,train_labels,batch_size=100,epochs=10,verbose=2)
```
